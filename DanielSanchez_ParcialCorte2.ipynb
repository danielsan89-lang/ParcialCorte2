```python
# Se importa BeautifulSoup
from bs4 import BeautifulSoup
import requests
```

```python
# Se pasa la url de la pagina a la que le vamos a consultar los datos html y traemos toda su estructura html
url = 'https://historiabreve.com/presidentes-de-colombia/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
soup
```

```python
# Se declara una variable que va a contener todo lo que sea table del html
tabla_presidentes = soup.find('table')
tabla_presidentes
```

```python
#Se declara una lista con un for que va a ir almacenando lo que encuentre dentro de las etiquetas tr, al final se imprime lo que hay dentro de esta lista.
headers = []

for a in tabla_presidentes.find_all('tr'):
    title = a.text.strip()
    headers.append(title)

headers
```
